```python
import numpy as np
import time
import math
n = int(input())
def jacobi (A , f , x ):
    xnew = n * [0]
    for i in range(n):
        s = 0
        for j in range(0, i):
            s = s + A [ i ][ j ] * x [ j ]
        for j in range(i + 1, n):
            s = s + A [ i ][ j ] * x [ j ]
        xnew[ i ] = ( f [ i ] - s ) / A [ i ][ i ]
    return xnew

def diff(x, xnew):
    s = float(0)
    for i in range(n):
        s += float((x[i] - xnew[i]) ** 2) 
    return float(math.sqrt(s))

def solve (A , f):
    eps = 0.01
    xnew = np.zeros(n)
    while True:
        x = xnew
        xnew = jacobi (A , f , x )
        if ( diff(x, xnew ) < eps ):
            return x


A = np.random.rand(n, n)
f = np.random.rand(n)
for i in range(n):
    for j in range(n):
        if i != j:
            A[i][i] += A[i][j]
print("My algorithm:")
my_time = time.time()
solve(A, f)
my_time = time.time() - my_time
for el in solve(A, f):
    print(el)
linalg_time = time.time()
print("Linalg:")
for el in np.linalg.solve(A, f):
    print(el)
linalg_time = time.time() - linalg_time
print("My time:", my_time)
print("Linalg time:", linalg_time)
```

```python
import numpy as np
import math
n = int(input())
def seidel (A , f , x ):
    xnew = np.zeros(n)
    for i in range(n):
        s = float(0)
        for j in range(i):
            s = float(s + A [ i ][ j ] * xnew [ j ])
        for j in range(i + 1, n):
            s = float(s + A[ i ][ j ] * x[ j ])
        xnew[ i ] = float(( f[ i ] - s) / A [ i ][ i ])
    return xnew

def diff(x, xnew):
    s = float(0)
    for i in range(0, n):
        s += float((x[i] - xnew[i]) ** 2)
    return float(math.sqrt(s))

def solve (A , f ):
    xnew = np.zeros(n)
    eps = 0.0000001
    while True:
        x = np.array(xnew)
        xnew = seidel (A , f , x )
        if(abs(diff(x, xnew)) < abs(eps)):
            break
    return x

A = np.random.rand(n, n)
f = np.random.rand(n)
for i in range(n):
    for j in range(n):
        if i != j:
            A[i][i] += A[i][j]
print(A)
print(f)
print("My algorithm:")
print(solve(A, f))
print("Linalg", np.linalg.solve(A, f))
```
